```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as se
from sklearn.cluster import KMeans
path = "/content/drive/MyDrive/Colab Notebooks/Mall_Customers.csv"
df = pd.read_csv(path)
```

```python
print(df.head())
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
X = df.iloc[:, [3, 4]].values
wcss = []
for i in range(1, 11):
    # init='k-means++' giúp chọn tâm cụm thông minh hơn để hội tụ nhanh
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('The Elbow Method')
plt.xlabel('Số lượng cụm (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
# Huấn luyện mô hình
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cụm 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cụm 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cụm 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cụm 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cụm 5')
# Vẽ các tâm cụm (Centroids)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Tâm cụm')
plt.title('Phân khúc khách hàng')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
